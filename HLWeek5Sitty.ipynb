```python
import altair as alt
```

```python
import pandas as pd
```

```python
data = pd.read_csv("Absenteeism_at_work.csv")
print (data)
```

```python
# Data Dashboard
```

```python
import altair as alt
alt.Chart(data).mark_point().encode(x= 'Distance from Residence to Work', y= 'Absenteeism time in hours'). interactive()
```

```python
brush = alt.selection_interval()
```

```python
points = alt.Chart(data). mark_point(). encode(x= 'Distance from Residence to Work', y= 'Absenteeism time in hours', color=alt.condition(brush, 'Absenteeism time in hours:N', alt.value('green'))). add_selection(brush)

points
```

```python
bar = alt.Chart(data). mark_bar(). encode(x= 'ID', y= 'Distance from Residence to Work', color=alt.condition(brush, 'Distance from Residence to Work:N', alt.value('red'))). add_selection(brush)

bar
```

```python
bars = alt.Chart(data). mark_bar(). encode(y= 'ID', color= 'ID:N', x='count(ID):Q'). interactive(). add_selection(brush)

bars
```

```python
bars = alt.Chart(data). mark_bar(). encode(y= 'ID', color= 'ID:N', x='count(ID):Q'). transform_filter(brush)

points & bars & bar
```

My chosen data file contained quantitative data and so i decided to first of all use a scatterplot as i wanted to check whether there would be any correlation between travel distance into work and the amount of absenteeism hours recorded.

I then went on to run a couple of bar charts fisrt to check which employee IDs had the highest count of absenteeism recorded and then check how far these employees resided from the place of work.

I decided on using colour in order for ease of reading and also made the charts interactive . 

My hypotethesis had been that there would be a correlation between the amount of absent hours recorded and the travel distance to work.
The data suggest this not to be the case as a majority of the amount of hours recorded were accounted to those residing within 25miles of the work place.
